```julia
using Distributions
using StatsBase: ecdf
using StatsPlots
default(fmt=:png)

function make_ecdf(X)
    ecdf_X = ecdf(X)
    f(x) = ecdf_X(x)
    f
end
```

```julia
dist = Weibull(2, 10)
```

```julia
n = 300
plot()
for _ in 1:200
    X = rand(dist, n)
    f = make_ecdf(X)
    plot!(t -> 0.85*f(t), 0, 30; label="", lw=0.3, alpha=0.3)
end
plot!(xtick=[0:10; 12:2:30], ytick=0:0.1:1)
```

```julia
n = 100
PP = []
for i in 1:50
    P = plot()
    for j in 1:2
        X = rand(dist, n)
        f = make_ecdf(X)
        plot!(t -> 0.85*f(t), 0, 20; label="")#, lw=0.3, alpha=0.3)
    end
    plot!(xtick=0:2:30, ytick=0:0.1:1, tickfontsize=6)
    push!(PP, P)
end

plot(PP...; layout=(10, 5), size=(1000, 1400))
```
