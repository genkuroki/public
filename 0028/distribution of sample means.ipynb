```julia
using Distributions
using StatsPlots
default(fmt=:png, titlefontsize=10)

function plot_samplemeans(;
        dist = MixtureModel([Normal(), Normal(10, 1)], [0.95, 0.05]),
        n = 20,
        L = 10^5,
        bin = 100,
        kwargs...
    )

    @show μ, σ = mean(dist), std(dist)
    X = rand(dist, n, L)
    meanX = mean.(eachcol(X))
    histogram(meanX; norm=true, alpha=0.3, label="sample means", bin, kwargs...)
    plot!(Normal(μ, σ/√n); label="normal approx")
    title!("MixtureModel([Normal(), Normal(10, 1)], [0.95, 0.05]), n = $n")
end
```

```julia
plot_samplemeans(; dist = MixtureModel([Normal(), Normal(10, 1)], [0.95, 0.05]), n = 10)
```
