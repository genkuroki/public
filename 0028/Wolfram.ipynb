```
pendulum = {y''[x] + Sin[y[x]] == 0, y[0] == 2, y'[0] == 1};
solpendulum = NDSolve[pendulum, y, {x, -10, 10}]
Plot[y[x] /. solpendulum, {x, -10, 10}]
Plot[Evaluate[{y[x], y'[x], y''[x]} /. solpendulum], {x, -10, 10}]
ParametricPlot[Evaluate[{y[x], y'[x]} /. solpendulum], {x, -10, 10}]
```
