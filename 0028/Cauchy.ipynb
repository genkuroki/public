```julia
using Distributions
using StatsPlots
default(fmt=:png)
```

```julia
function plot_Cauchy_samplemean(n; L = 10^5)
    X = rand(Cauchy(), n, L)
    X̄ = mean.(eachcol(X))
    histogram(X̄[@.(-100 < X̄ < 100)]; norm=true, alpha=0.3, bin=1000, label="")
    plot!(Cauchy(), -15, 15; label="Cauchy()", ls=:dash, lw=1.5)
    plot!(; xlim=(-15, 15))
end
```

```julia
plot_Cauchy_samplemean(1)
```

```julia
plot_Cauchy_samplemean(2)
```
