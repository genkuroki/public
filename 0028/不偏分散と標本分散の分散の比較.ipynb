```julia
using Distributions
using StatsPlots
```

```julia
dist = Normal()
n = 10
L = 10^6
X = rand(dist, n, L)
S² = var.(eachcol(X))
stdS² = √mean(s² -> (s² - var(dist))^2, S²)
V = (n-1)*S²/n
seV = √mean(v -> (v - var(dist))^2, V)
@show dist n kurtosis(dist) stdS² seV stdS²/seV mean(@. abs(S² - var(dist)) < abs(V - var(dist)))

stephist(.√S²; norm=true, label="√ sample unbiased variance")
stephist!(.√V; norm=true, label="√ sample variance")
vline!([std(dist)]; label="√ true variance", ls=:dash)
```
