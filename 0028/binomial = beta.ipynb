```julia
using Distributions
using StatsPlots
```

```julia
n, k = 10, 5
plot(t-> cdf(Beta(k, n-k+1), t), 0, 1; label="")
plot!(t -> ccdf(Binomial(n, t), k - 1), 0, 1; label="", ls=:dash)
plot!(t-> ccdf(Beta(k+1, n-k), t), 0, 1; label="")
plot!(t -> cdf(Binomial(n, t), k), 0, 1; label="", ls=:dash)
```

```julia
n, k = 11, 6
plot(t->   cdf(Beta(k, n-k+1), t), 0, 1; label="")
plot!(t -> ccdf(Binomial(n, t), k - 1), 0, 1; label="", ls=:dash)
plot!(t -> ccdf(Beta(k, n-k+1), t), 0, 1; label="")
plot!(t ->  cdf(Binomial(n, t), k - 1), 0, 1; label="", ls=:dash)
```
