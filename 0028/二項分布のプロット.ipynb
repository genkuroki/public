```julia
x ⪅ y = x < y || x ≈ y

function pval(bin, k)
    p0 = pdf(bin, k)
    sum(pdf(bin, j) for j in support(bin) if pdf(bin, j) ⪅ p0)
end
```

```julia
plot(100:113, x -> pval(Binomial(113, 0.95), x); xtick=0:200, ytick=0:0.05:1, label="")
```

```julia
plot(100:113, x -> pval(Binomial(113, 0.95), x) ≥ 0.05; xtick=0:200, ytick=0:0.05:1, label="")
```

```julia
plot(100:113, x -> pval(Binomial(113, 0.99), x) ≥ 0.05; xtick=0:200, ytick=0:0.05:1, label="")
```

```julia
beta_quantile(n, p) = Beta(p*(n+1), (1-p)*(n+1))
```

```julia
plot(0.8:0.001:1, x -> pdf(beta_quantile(113, 0.95), x); label="")
```

```julia
plot(0.8:0.001:1, x -> ccdf(beta_quantile(113, 0.95), x); ytick=0:0.05:1, label="")
```

```julia
plot(0.95:0.001:1, x -> pdf(beta_quantile(113, 0.99), x); label="")
```
