```julia
using Plots
default(fmt=:png)
using Printf

f(a, x) = a*x + 9 - 3a

ts = range(-π/2, π/2, 1003)[2:end-1]
@gif for t in [fill(ts[1], 30); ts[1:10:end]; fill(ts[end], 30)]
    a = tan(t)
    plot(x -> f(a, x), 0, 6; ylim=(-1, 11), label="y = ax + 9 - 3a")
    plot!(xtick=0:6, ytick=-1:11)
    title!(@sprintf("a = %10.3f", a))
    hline!([0]; c=:black, label="")
    scatter!([(3, 9), (3, 0)]; msw=0, c=:black, label="")
end
```
