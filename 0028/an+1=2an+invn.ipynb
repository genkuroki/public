```julia
using Plots
default(fmt = :png, legend = :topleft)

function a(n)
    n ≤ 1 && return 1//1
    2a(n-1) + 1//(n-1)
end

f(n) = 2^(n-1)*(1 + log(2))
```

```julia
[a.(1:10) f.(1:10)]
```

```julia
plot(a.(1:5); label="a_n")
plot!(f.(1:5); label="2ⁿ⁻¹(1 + log 2)", ls=:dash)
```
