$x^5+ax+b=0$ の解の1つを

$$x = -(b _4 F_3(1/5, 2/5, 3/5, 4/5 ;1/2, 3/4, 5/4 ;-(3125 b^4)/(256 a^5)))/a$$

で作れる.

```julia
using HypergeometricFunctions

f(a, b, x) = x*(x^4 + a) + b

# x = -(b _4 F_3(1/5, 2/5, 3/5, 4/5 ;1/2, 3/4, 5/4 ;-(3125 b^4)/(256 a^5)))/a
x1(a, b) = -(b/a) * pFq([1//5, 2//5, 3//5, 4//5], [1//2, 3//4, 5//4], -(3125//256)*(b/a)^4/a)

function test_x1(a, b)
    @show a, b
    @show -(3125//256)*(b/a)^4/a
    @show x = x1(a, b)
    @show f(a, b, x)
end
```

```julia
test_x1(3, 1);
```

```julia
test_x1(big(3), big(1));
```

```julia
test_x1(1, 1);
```

```julia
test_x1(big(1), big(1));
```

```julia
test_x1(1, 3);
```

```julia
test_x1(big(1), big(3));
```

```julia
for i in 1:20
    z = test_x1(randn(BigFloat), randn(BigFloat))
    if abs(z) > 0.01
        println("誤差が大きい.")
    end
    println()
end
```
