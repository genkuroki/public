```julia
using Random
Random.seed!(4649373)
using Plots

P(x, p) = p^x * (1 - p)^(1 - x)
L(X, p) = prod(P(x, p) for x in X)

@show X = rand(0:1, 12)
plot(p -> L(X, p), 0, 1; label="likelhood", xguide="p")
```

```julia
using Random
Random.seed!(4649373)
using Plots

sufficient_statistic(X) = (length(X), sum(X))

function likelihood(X, p)
    n, k = sufficient_statistic(X)
    p^k * (1 - p)^(n - k)
end

@show X = rand(0:1, 12)
plot(p -> likelihood(X, p), 0, 1; label="likelhood", xguide="p")
```

```julia
using BenchmarkTools

X = rand(0:1, 10^3)

a = @btime L($X, 0.5)
b = @btime likelihood($X, 0.5)
@show a b
```
