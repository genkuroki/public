```julia
using Plots

f(x, y) = 1  abs(abs(x) - 2) + abs(abs(y) - 2)  3
xs = -6:0.01:6
ys = -6:0.01:6
heatmap(xs, ys, f; aspectratio=:equal, c=:binary, colorbar=false)
plot!(xlim=extrema(xs), ylim=extrema(ys))
plot!(xguide="x", yguide="y")
title!("Region: 1  |x-2| + |y-2|  3")
plot!(size=(400, 400))
```
