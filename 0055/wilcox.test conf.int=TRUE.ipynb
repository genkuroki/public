```r
library(coin)
A = c(3.2, 5.5, 5.8, 6.0, 8.5)
B = c(1.1, 2.1, 3.1, 4.1)
df = data.frame(y=c(A, B), x=factor(c(rep("A", length(A)), rep("B", length(B)))))
wilcox_test(y ~ x, data=df, conf.int=TRUE, distribution="exact")
```
