<span style="font-size:16px;font-weight:bold">

* ゼロ主義(nullism)や二分法への執着(dichotomania)のような<span style="color:red">認知バイアス達</span>を「科学的原則」として神聖化し、数学的枠組みをあたかも物理的現実であるかのように扱い(モデルと現実の混同(reification))、確実性や最終性を渇望する人間のバイアスを無視することが、統計学の中核を腐敗させてきた。
* 解決策: 統計学を、確率論の一分野として<span style="border-bottom:solid 2px">ではなく</span>、認知科学と因果性理論を中核要素とする<span style="color:red">情報科学</span>として再構築する。
    
</span>

1 June 2022 Greenland – Reforming Statistics 4

------

<span style="font-size:16px">

DeFinettiの急進的ベイズ主義では、すべての確率は「主観的」であり、観察者の心の特性のみを記述する。その観点では

* パターンが「偶然によって引き起こされる」という考えは、世界についての因果的な言明としては馬鹿げている。
* むしろ我々は、妥当として提示される少数の因果的可能性の中から、非常に非ランダム(偏った)な選択を考慮することによって、認識されたパターンの<span style="font-weight:bold">因果的説明</span>を求める。
* そして我々は、考慮されなかった無数の因果的説明の残りを、「偶然」と呼ばれる形而上学的な原因を形成するものとしてモデルと現実を混同(reify)してしまう。

</span>

<span style="font-size:16px">

* <span style="font-weight:bold">すべての分析は、著しく不完全な感度分析の一部と見なされるべきである。</span>
* <span style="font-weight:bold">頻度論 vs ベイズの論争は、詳細な論理的分析の下では消滅する宗教論争である。</span>
* Boxian view: ベイズ的ツールは手法とモデル構築のため、頻度論ツールはそれらの評価のため(他にも多くの有用な組み合わせがある)。
* 最近まで、両方の「学派」は、その手法が導き出されるべき本質的な因果的／文脈的側面をカバーできていなかった。

<span style="font-size:16px">

<div align="center">
<span style="font-weight:bold">推論(INFERENCE)とは何か？</span>
</div>

* 辞書の例: <span style="font-weight:bold">「証拠と論理的思考に基づいて到達した結論。」</span>
* <span style="font-weight:bold;border-bottom:solid 2px">科学的推論</span><span style="font-weight:bold">とは、以下の仮定に基づいた、現実に関する複雑だが厳密に抑制された判断である: </span>

>論理的に一貫した「客観的」(観察者の外部にある)現実が存在し、発見可能な法則に従って我々の知覚を引き起こす: 
><div align="center"><span style="font-weight:bold">私の知覚 ← 現実 → あなたの知覚</span></div>

* これにより、推論は認知科学の一部となる。

</span>

1 June 2022 Greenland – Reforming Statistics 9

------

<span style="font-size:16px">

<div align="center">
<span style="font-weight:bold">科学的推論を以下と対比させる</span>
</div>

* <span style="font-weight:bold">「統計的推論」</span>は、あらゆる形式主義、「学派」、またはツールキットにおいて、<span style="font-weight:bold">データ処理プログラム (学習アルゴリズム) からの出力を受け取り、文脈から切り離されたルールを介して「推論」を生成するものとなっている。</span>
* それは、データを生成するメカニズム(データの<span style="font-weight:bold">原因</span>)の過度に単純化されたモデルを、抽象的な確率分布に変換する。
* <span style="font-weight:bold">それが残す意味論的な空白は、推論上の誤りを引き起こし、欺瞞(自己または他者に対する)を助長する。</span>

</span>

1 June 2022 Greenland – Reforming Statistics 10

------

<span style="font-size:16px">

* <span style="font-weight:bold">統計学は意味論や日常言語を無視または軽蔑し、その代わりに、研究がそれに近いものを何も提供していない場合でさえ、相当な論理的飛躍なしには成り立たない「有意性」や「信頼」を約束する、欺瞞的な専門用語を好んだ。</span>
* これは、「ソフトサイエンス」のほとんどのユーザーや消費者がその仮定や危険性をほとんど理解していない、難解な形式主義、記法、<span style="color:red;font-weight:bold">見せかけの精度 (artificial precision)</span>に基づく技術製品やサービスを売り込むために行われた。<br>
<b>― 医療製品の販売との類似性に注目！</b>

* <span style="font-weight:bold">二分法は、決定的な結論を求める人間の欲求を満たす。</span>なぜなら、たとえ研究(実際の物理的なデータ生成装置)が、批判的に精査されればそのような結論を強制することができなくても、二分法は適用されるからである。\*

\*「さらなる研究が必要である」という結論は別として。ただし、多くの場合、費用便益の考慮や他の研究に照らして、それさえも正当化されない。

<span style="font-size:16px">

<b>典型的な例</b>(Brown et al., “Association between serotonergic antidepressant use during pregnancy and autism spectrum disorder in children” (妊娠中のセロトニン作動性抗うつ薬\[SSRI\]使用と子供の自閉症スペクトラム障害\[ASD\]との関連), JAMA 2017;317:1544-52)、要旨: 

* 「\[Coxモデル\]調整済みHR、<span style="color:red;font-weight:bold">1.59</span> \[95% CI, <span style="color:red;font-weight:bold">1.17, 2.17</span>\]。IPTW HDPS後、関連は有意ではなかった(HR, <span style="color:red;font-weight:bold">1.61</span> \[95% CI: <span style="color:red;font-weight:bold">0.997, 2.59</span>\])。」(<span style="font-weight:bold">p = 0.0505</span>)
* 彼らの結論: 「<b>子宮内での曝露は、自閉症スペクトラム障害と<span style="color:red">関連していなかった</span></b>」

<b>訳註:</b>
* IPTW (Inverse Probability of Treatment Weighting)：治療割り当ての逆確率で重み付けして交絡を調整する方法。
* HDPS (High-Dimensional Propensity Score)：高次元の交絡因子を用いた傾向スコア補正。
* HR (Hazard Ratio)：ハザード比。1より大きければ曝露がリスク上昇に関連することを示唆。
* p = 0.0505：有意差とみなすことが多い閾値 α = 0.05 をわずかに超えている。
