```julia
using Distributions
using StatsPlots
default(size=(400, 300), titlefontsize=10, tickfontsize=6)
plot(sin);
```

```julia
zstat(dist::UnivariateDistribution, X) = zstat(mean(dist), std(dist), X)
chisqstat(dist::UnivariateDistribution, X) = chisqstat(var(dist), X)

function zstat(μ, σ, X)
    X̄, n = mean(X), length(X)
    Z = (X̄ - μ)/(σ/√n)
end

function chisqstat(σ², X)
    U², n = var(X), length(X)
    C = (n - 1)*U²/σ²
end
```

```julia
dist = Normal(2, 3)
n = 10
L = 10^5
X = [rand(dist, n) for _ in 1:L]
Z = zstat.(dist, X)
C = chisqstat.(dist, X)

P = histogram(Z; norm=true, alpha=0.3, label="")
plot!(Normal(); label="")

Q = histogram(C; norm=true, alpha=0.3, label="")
plot!(Chisq(n - 1); label="")

plot(P, Q; size=(800, 300))
```
