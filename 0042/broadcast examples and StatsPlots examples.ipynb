```julia
using Plots

n = 100
x = range(-1, 1, n)
u = 0.5randn(n)
y = sinpi.(x) + u

d = 3
A = x .^ (0:d)'
 = A \ y
f(x) = evalpoly(x, )

scatter(x, y; label="data")
plot!(f; label="degree-$d polynomial")
```

```julia
using Distributions
using StatsPlots

n = 10^6
X = sort!(randn(n))
Y = X + reverse(X)
@show normal = fit(Normal, Y)
stephist(Y; norm=true, label="data")
plot!(normal; label="normal approx.")
```

```julia
using Distributions
using StatsPlots

n = 10^6
X = sort!(randn(n))
@views Y = X[1:2:end] + X[2:2:end]
@show normal = fit(Normal, Y)
stephist(Y; norm=true, label="data")
plot!(normal; label="normal approx.")
```

```julia
using Distributions
using StatsPlots

n = 10^6
X = sort!(randn(n))
@views Y = X[1:end2] + X[end2+1:end]
@show normal = fit(Normal, Y)
stephist(Y; norm=true, label="data")
plot!(normal; label="normal approx.")
```
