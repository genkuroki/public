```julia
using Plots
default(fmt=:png, titlefontsize=12)

harmonic_s(N) = sum(1/k for k in 1:N)
approx_hs0(N) = log(N) + MathConstants.eulergamma
approx_hs1(N) = log(N) + MathConstants.eulergamma + 1/(2N)
approx_hs2(N) = log(N) + MathConstants.eulergamma + 1/(2N) - 1/(12N^2)

@show hs = sum(1//k for k in 1:10)
@show float(hs)
println()
@show harmonic_s(10) approx_hs2(10) approx_hs1(10) approx_hs0(10);

Nmax = 15
N = 1:Nmax
scatter(N, harmonic_s.(N); label="1/1 + 1/2 +  + 1/N")
plot!(approx_hs1, 0.01, Nmax+0.7; label="log(x) +  + 1/(2x)")
plot!(approx_hs0, 0.01, Nmax+0.7; label="log(x) + ", ls=:dash)
plot!(xguide="N, x", ylim=(-0.2, approx_hs1(Nmax+1)))
title!("1/1 + 1/2 +  + 1/N  log(x) +  + 1/(2N)")
```
