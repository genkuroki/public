```julia
ENV["LINES"] = 200
ENV["COLUMNS"] = 1000

function f(n)
    A = Vector{Int}(undef, n-1)
    y = fill(1, n-1)
    for k in 1:n-1
        y .*= 1:n-1
        y .= mod.(y, n)
        A[k] = sum(y)
    end
    A
end
```

```julia
[p*(p-1)÷2 for p in (2, 3, 5, 17, 257, 65537)]
```

```julia
@show m = 2^0
@show n = m + 1
@time A = f(n)
@show A
@show unique(A[setdiff(1:n-1, [m])])
@show A[m];
```
