$$
\newcommand\aic{\mathrm{aic}}
\newcommand\lik{\mathrm{lik}}
\aic_0 = -2\lik_0, \qquad
\aic_1 = -2\lik_1 + 2
\aic_0 - \aic_1 = 2(\lik_1 - \lik_0) - 2, \quad
$$

$$
2(\lik_1 - \lik_0) > c \iff \aic_0 - \aic_1 > c - 2
$$

```julia
using Distributions
quantile(Chisq(1), 0.95) - 2
```
