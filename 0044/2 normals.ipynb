```julia
using Distributions
using StatsPlots
default(fmt=:png)
```

```julia
α = 0.05
c = quantile(Normal(), 1 - α/2)
```

```julia
plot(Normal())
plot!(Normal(2c))
```

```julia
d = quantile(Normal(0, √2), 1 - α/2)
```

```julia
plot(Normal())
plot!(Normal(d))
```

```julia
2ccdf(Normal(), d - c)
```
