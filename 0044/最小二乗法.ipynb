```julia
using Distributions
using LinearAlgebra
norml2(x) = dot(x, x)
using StatsPlots
default(fmt=:png)

@show β = [10, 5]
@show σ = 50

n = 100
x = rand(Normal(10, 10), n)
y = evalpoly.(x, (β,)) + σ*randn(n)

X = x .^ (0:1)'
@show β̂ = X \ y
ŷ = evalpoly.(x, (β̂,))
@show ŝ = √(norml2(y - ŷ) / (n - 2))

xs = range((extrema(x) .+ (-5, 5))..., 1000)
yhats = evalpoly.(xs, (β̂,))

scatter(x, y; label="data")
plot!(xs, yhats; label="regression line")
```
