```julia
using Distributions
using StatsPlots
default(fmt=:png)
```

```julia
n = 1000
x = rand(Normal(10, 2), n)
y = 5 .+ x .+ 3rand(Normal(), n)

A = x .^ (0:1)'
β̂ = A \ y

scatter(x, y; label="data (xᵢ, yᵢ)", ms=3, msc=:auto, alpha=0.7)
plot!(x -> evalpoly(x, β̂); label="", lJJw=2)
plot!(xguide="x", yguide="y")
plot!(size=(400, 400))
```

```julia
ŷ = evalpoly.(x, (β̂,))
u = y - ŷ
scatter(x, u; label="", title="(xᵢ, uᵢ)", ms=3, msc=:auto, alpha=0.7)
hline!([0]; label="")
plot!(xguide="x", yguide="u")
plot!(size=(400, 400))
```

これだと, $x$ の値が $x=10$ から離れると $u$ の分布がまばらになっているかのように誤解してしまうかもしれない.

```julia
idx_sorted = sortperm(x)
u_sorted = u[idx_sorted]

scatter(1:n, u_sorted; label="", title="sorted (i, uᵢ)", ms=3, msc=:auto, alpha=0.7)
hline!([0]; label="")
plot!(size=(400, 400))
```

$x$ の値に $u$ の分布がよってなさそうに見える.

```julia
n = 1000
x = rand(Normal(), n)
y = exp.(0.2 .+ 0.3x .+ 0.3rand(Normal(), n))

A = x .^ (0:1)'
β̂ = A \ y

scatter(x, y; label="data (xᵢ, yᵢ)", ms=3, msc=:auto, alpha=0.7)
plot!(x -> evalpoly(x, β̂); label="", lw=2)
plot!(xguide="x", yguide="y")
plot!(size=(400, 400))
```

```julia
ŷ = evalpoly.(x, (β̂,))
u = y - ŷ
scatter(x, u; label="", title="(xᵢ, uᵢ)", ms=3, msc=:auto, alpha=0.7)
hline!([0]; label="")
plot!(xguide="x", yguide="u")
plot!(size=(400, 400))
```

$x$ の値が大きくなると $u$ の分散が大きくなって行っているように見える.
