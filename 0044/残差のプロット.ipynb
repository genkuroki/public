```julia
using Distributions
using StatsPlots
default(fmt=:png)
```

```julia
n = 1000
x = rand(Normal(10, 2), n)
y = 5 .+ x .+ 3rand(Normal(), n)

A = x .^ (0:1)'
 = A \ y

scatter(x, y; label="data (x, y)", ms=3, msc=:auto, alpha=0.7)
plot!(x -> evalpoly(x, ); label="", lJJw=2)
plot!(xguide="x", yguide="y")
plot!(size=(400, 400))
```

```julia
y = evalpoly.(x, (,))
u = y - y
scatter(x, u; label="", title="(x, u)", ms=3, msc=:auto, alpha=0.7)
hline!([0]; label="")
plot!(xguide="x", yguide="u")
plot!(size=(400, 400))
```

, $x$  $x=10$  $u$ .

```julia
idx_sorted = sortperm(x)
u_sorted = u[idx_sorted]

scatter(1:n, u_sorted; label="", title="sorted (i, u)", ms=3, msc=:auto, alpha=0.7)
hline!([0]; label="")
plot!(size=(400, 400))
```
