```julia
using Plots

function pitick(start, stop, denom)
    a = Int(cld(start, /denom))
    b = Int(fld(stop, /denom))
    tick = range(a*/denom, b*/denom; step=/denom)
    ticklabel = piticklabel.(a:b, denom)
    tick, ticklabel
end

function piticklabel(x, denom)
    d = repr(denom)
    x == 0 && return "0"
    x == 1 && return "/" * d
    x == -1 && return "-/" * d
    if mod(x, denom) == 0
        q = x  denom
        q == 1 && return ""
        q == -1 && return "-"
        return repr(q) * ""
    end
    return repr(x) * "/" * d
end

a, b = -2, 2
plot(sin, a, b; xtick=pitick(a, b, 4), label="y = sin(x)", size=(720, 300))
```
