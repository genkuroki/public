```julia
function pdfpseudot(t, κ; n = 10, trange = (-Inf, Inf))
    ε = κ/n
    η = 2/(n - 1)
    θ = ε + η
    λ = ε / θ
    g(y) = pdf(Gamma(1/θ, θ), y)
    f(t, y) = pdf(truncated(Normal(0, 1/y^(η/θ/2)), trange...), t)
    F(y) = f(t, y) * g(y) 
    quadgk(F, 0, Inf)[1]
end

function pdffalset(t, κ; n = 10)
    f(z) = pdf(Normal(), z)
    ε = κ/n
    η = 2/(n - 1)
    θ = ε + η
    λ = ε / θ
    g(y) = pdf(Gamma(1/θ, θ), y)
    F(y) = f(t/y) / y * g(y) 
    quadgk(F, 0, Inf)[1]
end

function cdfpseudot(t, κ; n = 10, trange = (-Inf, Inf))
    ε = κ/n
    η = 2/(n - 1)
    θ = ε + η
    λ = ε / θ
    g(y) = pdf(Gamma(1/θ, θ), y)
    f(t, y) = cdf(truncated(Normal(0, 1/y^(η/θ/2)), trange...), t)
    F(y) = f(t, y) * g(y) 
    quadgk(F, 0, Inf)[1]
end
```

```julia
@show κ = -2
@show n = 10
@show ε = κ/n
@show η = 2/(n - 1)
@show θ = ε + η
@show α = 1 / θ
@show λ = ε / θ
@show η/θ

trange = (-Inf, Inf)
@show 2(1 - cdfpseudot(quantile(TDist(n-1), 0.975), κ; n, trange))

plot(t -> pdfpseudot(t, κ; n, trange), -6, 6; label="pseudo t")
plot!(TDist(n-1), -6, 6; label="TDist($n-1)", ls=:auto)
#plot!(Normal(), -6, 6; label="Normal()", ls=:auto)
#plot!(t -> pdffalset(t, κ; n), -6, 6; label="false t", ls=:auto)
```

```julia
conditioned_ecdf(z, y, Z, Y; δ = std(Y)/100) = ecdf(Z[y - δ .< Y .< y + δ])(z)
conditioned_mean(y, Z, Y; δ = std(Y)/100) = mean(Z[y - δ .< Y .< y + δ])
```

```julia
dist = Normal()
@show n = 10
@show L = 10^7
@show μ, σ = mean(dist), std(dist)
@show κ = kurtosis(dist)
@show ε = κ/n
@show η = 2/(n-1)
@show θ = ε + η
@show s = √θ

X = [rand(dist, n) for _ in 1:L]
X̄ = mean.(X)
Z = @. √n*(X̄ - μ)/σ
Y = var.(X) / σ^2

@show 2(1 - cdfpseudot(quantile(TDist(n-1), 0.975), κ; n))
@show 1 - ecdf(Z.^2 ./ Y)(quantile(TDist(n-1), 0.975)^2)

P = plot()
for (k, y) in enumerate(round.((exp(-2s), exp(-s), 1, 1+s, 1+2s); digits=3))
    @show conditioned_mean(y, Z.^2, Y)
    plot!(z -> 2(1 - conditioned_ecdf(z, y, Z, Y)), 0, 5; label="$y", c=k)
    plot!(z -> 2ccdf(Normal(0, √(y^(ε/θ))), z), 0, 5; label="$y", c=k, ls=:dash)
end
plot!()
```
