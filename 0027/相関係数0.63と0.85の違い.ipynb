```julia
using Distributions
using Plots
```

```julia
dist = MvNormal([0,0], [1 0.63; 0.63 1])

PP = []
k = 0
while true
    X = rand(dist, 30)
    c = cor(X[1,:], X[2,:])
    if 0.62 < c < 0.64
        P = scatter(X[1,:], X[2,:]; label="", xlim=(-3, 3), ylim=(-3, 3), title="$c", msc=:auto)
        push!(PP, P)
        k += 1
    end
    k ≥ 25 && break
end
plot(PP...; layout=(5, 5), size=(1000, 1000), titlefontsize=8, tickfontsize=6, ms=3, fmt=:png)
```
