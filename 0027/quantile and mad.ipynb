```julia
using Distributions
using Roots
using StatsPlots
```

```julia
dist = Gamma(1.1, 1)
μ = mean(dist)
Q = quantile.(dist, 0.25:0.25:0.75)
a = quantile.(dist, 0.5)
d = find_zero(d -> cdf(dist, a+d) - cdf(dist, a-d) - 0.5, 0)
plot(dist; label="dist")
vline!([μ], label="mean")
vline!(Q; label="quantile", ls=:dash)
vline!([a-d, a, a+d]; label="mad", ls=:dot, lw=1.5)
plot!(xlim=(-0.1, 8))
```
