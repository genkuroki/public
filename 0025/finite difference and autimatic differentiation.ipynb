```julia
using Plots

f(x) = x^3 - 3x^2 + 3x - 3
df(x) = 3x^2 - 6x + 3

x = 2.1
dx = @. 2.0 ^ (-60:4)

finitediff = @. (f(x + dx) - f(x)) / dx
abserr_finirediff = @. abs(finitediff - df(x))

plot(dx, abserr_finirediff; xscale=:log10, yscale=:log10, label="")
plot!(; xtick=@.(1e1^(-20:2:2)), ytick=@.(1e1^(-20:2:2)))
plot!(; xlabel="dx", ylabel="absolute error")
```
