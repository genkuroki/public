```julia
using RCall
@rimport plotrix as plotrix
```

```julia
x = [21, 62, 10, 53]
labels = ["London", "New York", "Singapore", "Mumbai"]
plotrix.pie3D(x; labels)
```

```julia
plotrix.pie3D(x; labels, radius=1.6, theta=π/5, start=π/2)
```

```julia
y = [13, 20, 18, 9, 4, 27, 9]
labels = string.(10:10:70) .* "歳"
plotrix.pie3D(reverse(y); labels=reverse(labels), radius=1.4, theta=π/7, start=π/2)
```
