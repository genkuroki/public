```julia
using Distributions
using StatsPlots
default(fmt=:png)

,  = 20, 180
,  = 10, 190
beta1 = Beta(, )
beta2 = Beta(, )

L = 10^6
P = rand(beta1, L)
Q = rand(beta2, L)
logP = @. log(P)
logQ = @. log(Q)
logRR = logP - logQ
normal = Normal(log(/(+)) - log(/(+)), (//(++1) + //(++1)))

density(logRR; label="density of log Beta($, $)  log Beta($, $)")
plot!(normal; label="normal approximation", ls=:dash)
plot!(legend=:outertop, legendfontsize=12)
```
