```julia
using Distributions
#using StatsPlots
using Plots
default(fmt=:png)

n = 10^4
L = 100
X = rand(TDist(2), n, L)
Y = cumsum(X; dims=1) ./ (1:n)
ns = [1:10; n100:n100:n]
plot(ns, Y[ns,:]; label="", lw=0.3, alpha=0.6, ylim=(-1, 1))
plot!(n -> 6/n; label="6/n", c=:red)
plot!(n -> -6/n; label="", c=:red)
```
