```julia
using Distributions
using QuadGK
using Plots
default(fmt=:png)

f(x) = quadgk(t -> exp(-t^2), 0, x)[1]
g(x) = sin(sin(x))

plot(f, -/2, /2; label="\$\\int_0^x \\exp(-t^2)\\,dt\$")
plot!(g; label="\$\\sin(\\sin(x))\$", ls=:dash)
plot!(xguide="\$x\$", guidefontsize=16)
plot!(legendfontsize=16)
```
