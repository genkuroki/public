```julia
# https://twitter.com/iitenki_moruten/status/1434070751884177415

using Plots
using StatsBase
using Random
using Printf
using Distributions
using BenchmarkTools
using QuadGK
using SpecialFunctions
using Zygote

# Generating Gaussian distribution (1-dim) by using HMC method
# 

v(x) = 0.5*x^2
u(x) = x^2*(x-3)^2/2

function S(x,n_size)
    sum(u(x[i]) for i=1:n_size)
end

function dSdx(x,k)
    return u'(x) 
end

function H(x,p,n_size)
    K = sum(p[i]^2 for i=1:n_size)
    return S(x,n_size) + 0.5*K
end

#HMC
function HMC_1dim(x,parameter)
        n_size,n_tau, = parameter
        #
        p_ini = rand(Normal(0,1),n_size)
        p_fin = zeros(n_size)
        x_ini = zeros(n_size)
        x_fin = zeros(n_size)
    
        #
        for i=1:n_size x_ini[i] = x[i] end
    
        #
        H_ini = H(x_ini,p_ini,n_size)
    
        #
        leapfrog!(x,p_ini,parameter)
    
        #
        for i=1:n_size x_fin[i] = x[i]; p_fin[i] = p_ini[i] end
    
        #
        H_fin = H(x_fin,p_fin,n_size)
    
        #Metropolis check
        r=rand()
        if r<exp(H_ini - H_fin)
            for i=1:n_size x[i] = x_fin[i] end
        else
            for i=1:n_size x[i] = x_ini[i] end
        end
end

#x,p
function leapfrog!(x,p,parameter)
    n_size,n_tau, = parameter
      = /n_tau 
    p0  = 0.0
    p1  = 0.0
    x05 = 0.0
    x15 = 0.0
    for k=1:n_size
        #initial setting : 0
        p0  = p[k]
        x05 = x[k] + p0 * 0.5*
    
        #(n-1)step : 0 -> n-1 
        for i=1:n_tau-1
            p1  = p0 - dSdx(x05,k)*
            x15 = x05 + p1*
            p0  = p1
            x05 = x15
        end
    
        #final step : n-1 -> n
        p_end = p1 - dSdx(x15,k)*
        x_end = x15 + p_end*0.5*
        
        x[k]  = x_end
        p[k]  = p_end
    end
end

function gen_configs(,n_con,parameter)
    n_size,n_tau, = parameter
    x = zeros(n_size)
    for j=1:n_con
        HMC_1dim(x,parameter)
        [j] = x[1]
    end
end

##########################################################

"""q = [n_size,n_tau,]"""

#
n_config = 1_000_000          # # of configurations
n_size   = 1
n_tau    = 100
        = 5
q   = [n_size,n_tau,]
        = zeros(n_config)

#result = @benchmark gen_configs(n_config,q)
@time gen_configs(,n_config,q)

##########################################################


###plot###
gauss(x) = exp(-0.5*x^2)/sqrt(2*pi)
pdf_u(x) = exp(-u(x))
Z0,error = quadgk(pdf_u,-Inf,Inf)
normed_u(x) = pdf_u(x)/Z0

xr = range(-3,6,length=1000)

plot(,st=:histogram,nbins=50,norm=:pdf,alpha=0.3,label="HMC's sample",legend=:topleft)
plot!(xr,normed_u.(xr),label="pdf_ana")
```

```julia
using Plots
# using StatsBase
# using Random
# using Printf
using Distributions
using BenchmarkTools
using QuadGK
# using SpecialFunctions

module Revised

using Random
using Distributions
using ForwardDiff

# Generating Gaussian distribution (1-dim) by using HMC method
# 

S(x, parameter) = ((u,) = parameter; sum(u, x))
dSdx(x, k, parameter) = ((u,) = parameter; ForwardDiff.derivative(u, x))
square(x) = x^2
H(x, p, parameter) = S(x, parameter) + 0.5sum(square, p)

#HMC
function HMC_1dim!(x, parameter)
        _, n_tau, , p_ini, p_fin, x_ini, x_fin = parameter
        
        #
        rand!(Normal(0,1), p_ini)
    
        #
        x_ini .= x
    
        #
        H_ini = H(x_ini, p_ini, parameter)
    
        #
        leapfrog!(x, p_ini, parameter)
    
        #
        x_fin .= x
        p_fin .= p_ini
    
        #
        H_fin = H(x_fin, p_fin, parameter)
    
        #Metropolis check
        r = rand()
        if r  min(1, exp(H_ini - H_fin))
            x .= x_fin
        else
            x .= x_ini
        end
end

#x,p
function leapfrog!(x, p, parameter)
    n_size = length(x)
    _, n_tau,  = parameter
      = /n_tau
    for k=1:n_size
        #initial setting : 0
        p0  = p[k]
        x05 = x[k] + p0 * 0.5*
    
        #(n-1)step : 0 -> n-1
        local p1, x15
        for i=1:n_tau-1
            p1  = p0 - dSdx(x05, k, parameter)*
            x15 = x05 + p1*
            p0  = p1
            x05 = x15
        end
    
        #final step : n-1 -> n
        p_end = p1 - dSdx(x15, k, parameter)*
        x_end = x15 + p_end*0.5*
        
        x[k]  = x_end
        p[k]  = p_end
    end
end

function gen_configs(, x, parameter)
    n_tau,  = parameter
    for j in eachindex()
        HMC_1dim!(x, parameter)
        [j] = x[1]
    end
end

end

v(x) = 0.5x^2
u(x) = x^2*(x-3)^2/2

##########################################################

n_config = 10^6          # # of configurations
n_size   = 1
n_tau    = 100
        = 5

        = zeros(n_config)
x        = zeros(n_size)
p_ini    = similar(x)
p_fin    = similar(x)
x_ini    = similar(x)
x_fin    = similar(x)

"""q = (u, n_tau, , p_ini, p_fin, x_ini, x_fin)"""
q   = (u, n_tau, , p_ini, p_fin, x_ini, x_fin)

#result = @benchmark gen_configs(n_config,q)
@time Revised.gen_configs(, x, q)

##########################################################

###plot###
gauss(x) = exp(-0.5*x^2)/sqrt(2*pi)
pdf_u(x) = exp(-u(x))
Z0,error = quadgk(pdf_u,-Inf,Inf)
normed_u(x) = pdf_u(x)/Z0

xr = range(-3,6,length=1000)

plot(,st=:histogram,nbins=50,norm=:pdf,alpha=0.3,label="HMC's sample",legend=:topleft)
plot!(xr,normed_u.(xr),label="pdf_ana")
```
