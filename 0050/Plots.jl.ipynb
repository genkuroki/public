```julia
# JuliaLang

using Plots
default(fmt=:png, framestyle=:zerolines,
    legend=false, tick=false, size=(400, 400),
    lw=2.5, c=:red)

P1 = plot([range(-7, -eps(), 500); NaN; range(eps(), 7, 500)], x -> 1/x)
plot!(ylim=(-7, 7))
title!("y = 1/x")

P2 = plot(x -> -x^2+5, -3, 3)
plot!(xlim=(-4, 4), ylim=(-8, 8))
title!("y = -x² + 5")

P3 = plot(range(-6, 6, 1000), x -> x^3-3x^2+4)
plot!(ylim=(-6, 6))
title!("y = x³ - 3x² + 4")

P4 = plot(range(eps(), 1000, 1000), x -> x > 0 ? log10(x) : NaN; lw=3)
plot!(xlim=(-200, 1000), ylim=(-1, 10))
title!("y = log₁₀ x")

P5 = plot(range(-3, 15, 1000), x -> 2^x; lw=3)
plot!(xlim=(-3, 15), ylim=(-1, 20))
title!("y = 2ˣ")

P6 = plot(sin, -8, 8)
plot!(ylim=(-7, 7))
title!("y = sin x")

plot(P1, P2, P3, P4, P5, P6; size=(900, 600), layout=(2, 3))
```
