```julia
using StaticArrays
struct Point{T} <: FieldVector{2, T} x::T; y::T end
```

```julia
p = Point(1, 2)
```

```julia
q = Point(3, 4)
```

```julia
q - p
```

```julia
3p
```

```julia
using LinearAlgebra
dot(p, q), cross(p, q)
```
