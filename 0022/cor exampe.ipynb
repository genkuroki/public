```julia
using Statistics
using StatsPlots
using Random

Random.seed!(12345)

n = 20
x = randn(n)
y = randn(n)

@show X = round.(3x - 1.5y, digits=2)
@show Y = round.(-x + 2y, digits=2)
println()
@show cor(X, Y)

plot(1:n, [X Y]; label=["X" "Y"], xtick=1:20)
```

```julia
using LinearAlgebra

mX, mY = mean(X), mean(Y)
nX, nY = norm(X .- mX), norm(Y .- mY)
theta = acos(cor(X, Y))

plot([0, nX], [0, 0]; arrow=true, label="X - X")
plot!([0, nY*cos(theta)], [0, nY*sin(theta)]; arrow=true,label="Y - ")
plot!(; xlim=(-20, 20), ylim=(-20, 20), size=(400, 400))
title!("cos() = $(cos(theta))", titlefontsize=12)
```
