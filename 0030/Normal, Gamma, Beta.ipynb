```julia
using Distributions
using StatsPlots
default(fmt=:png)
```

```julia
L = 10^6
A = randn(L)
B = randn(L)
C = randn(L)
D = randn(L)
E = randn(L)
F = randn(L)
G = @. 2(A^2 + B^2 + C^2 + D^2 + E^2 + F^2)
```

```julia
stephist(G; norm=true, label="G")
plot!(Gamma(3, 4); label="Gamma(3,4)", ls=:dash)
```

```julia
X = rand(Gamma(3, 1), L)
Y = rand(Gamma(7, 1), L)
T = @. X/(X+Y)
```

```julia
stephist(T; norm=true, label="T")
plot!(Beta(3,7); ls=:dash, label="Beta(3,7)")
```
