```julia
ENV["LINES"] = 2^10
using PyCall

fm = pyimport("matplotlib.font_manager")
ttffonts = sort(fm.findSystemFonts(fontpaths=nothing, fontext="ttf"))
```

```julia
ipafonts = sort(filter(f->contains(f, r"IPA"i), ttffonts))
ipanames = sort((f->fm.FontProperties(fname=f).get_name()).(ipafonts))
```

```julia
jugemfonts = sort(filter(f->contains(f, r"JuGeM"i), ttffonts))
jugemnames = unique(sort((f->fm.FontProperties(fname=f).get_name()).(jugemfonts)))
```

```julia
yufonts = sort(filter(f->contains(f, r"Yu"), ttffonts))
yunames = unique(sort((f->fm.FontProperties(fname=f).get_name()).(yufonts)))
```
