```julia
using Distributions
using StatsPlots
default(fmt=:png)
```

```julia
n = 10
p = 5/n
dist = Binomial(n, p)
normal = Normal(mean(dist), std(dist))
bar(dist, 0:min(n, round(Int, 4n*p)); label="Binomial($n, $p)", alpha=0.3)
plot!(x -> pdf(Poisson(n*p), round(Int, x)); label="Poisson($(n*p))", ls=:dash)
plot!(normal; label="normal approx.", lw=1.5)
```
