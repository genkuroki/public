```julia
using Distributions

dist = Gamma(2, 3)

n = 10
@show x = rand(dist, n); # 標本の数値

@show x̅ = mean(x); # 標本平均
@show s² = var(x); # 不偏分散
@show t = (x̅ - 5) / √(s²/n); # t値
@show pval = 2ccdf(TDist(n-1), abs(t));
```

```julia
@show dist = Gamma(2, 3); # 仮想的な母集団分布
@show mean(dist); # 母平均
@show var(dist); # 母分散
@show skewness(dist); # 母歪度
@show kurtosis(dist); # 母尖度
```

```julia
using Distributions
using StatsPlots
default(fmt=:png)

dist = Gamma(2, 3)
plot(dist; label="dist", c=2)
```

```julia
n = 1000
X = rand(dist, n)
histogram(X; norm=true, label="sample", alpha=0.3)
```

```julia
poi = Poisson(10)
bar(poi; alpha=0.3, label="poi", c=1)
plot!(xlim=(-0.5, 25.5))
```

```julia
bin = Binomial(200, 0.05)
bar(bin; alpha=0.3, label="bin", c=2)
plot!(xlim=(-0.5, 25.5))
```

```julia
poi = Poisson(10)
bin = Binomial(200, 0.05)
bar(poi; alpha=0.3, label="poi", c=1)
bar!(bin; alpha=0.3, label="bin", c=2)
plot!(xlim=(-0.5, 25.5))
```

```julia
groupedbar([pdf.(poi, 0:25) pdf(bin, 0:25)]; alpha=0.3, label=["poi" "bin"])
```

```julia
poi = Poisson(10)
bin = Binomial(200, 0.05)
plot(x -> pdf(poi, round(x)), -0.5, 25.5; label="poi", c=1)
plot!(x -> pdf(bin, round(x)); label="bin", ls=:dash, c=2)
```

```julia
L = 10^3
X = rand(poi, L)
Y = rand(bin, L)
stephist(X; norm=true, bin=-0.5:25.5, label="X")
stephist!(Y; norm=true, bin=-0.5:25.5, label="Y", ls=:dash)
title!("L = $L")
```
