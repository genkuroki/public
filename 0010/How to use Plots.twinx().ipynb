```julia
using Plots
using Plots.PlotMeasures
pyplot(fmt = :svg)
bar(rand(10), c=:white, right_margin=1cm)
plot!(twinx(), rand(10), label="v1")
plot!(twinx(), rand(10) * 10, label="v2")
```
