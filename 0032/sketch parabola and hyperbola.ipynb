```julia
using Plots
default(fmt=:png)
using StatsFuns

X(x, y) = x/y
Y(x, y) = 1 - 1/y
```

```julia
x = range(-10, 10, 3000) .^5
y = @. (x+2)^2 + 2
xs = range(-10, 10, 10) .^5
ys = logit.(range(0.5, 1, 10))

plot(; framestyle=:none)
for a in -10:10
    plot!(X.(a, ys), Y.(a, ys); label="", c=2, lw=0.3)
end
for b in 3:3000
    plot!(X.(xs, b), Y.(xs, b); label="", c=2, lw=0.3)
end
plot!(X.(x,y), Y.(x, y); label="", c=1)
plot!(xlim=(-1.2, 0.2), ylim = (0.4, 1.2))
title!("y = (x + 2)² + 2")
```

```julia
x = range(-10, 10, 3000) .^5
y = @. √((x+2)^2 + 2)
xs = range(-10, 10, 10) .^5
ys = logit.(range(0.5, 1, 10))

plot(; framestyle=:none)
for a in -10:10
    plot!(X.(a, ys), Y.(a, ys); label="", c=2, lw=0.3)
end
for b in 3:3000
    plot!(X.(xs, b), Y.(xs, b); label="", c=2, lw=0.3)
end
plot!(X.(x,y), Y.(x, y); label="", c=1)
plot!(xlim=(-2.2, 2.2), ylim = (0, 2))
title!("y = √((x + 2)² + 2)")
```

```julia
x = range(-10, 10, 1000) .^5
y = @. √((x+2)^2 + 2)
xs = range(-10, 10, 10) .^5
ys = logit.(range(0.5, 1, 10))

plot(; framestyle=:none)
for a in -10:10
    plot!(X.(a, ys), Y.(a, ys); label="", c=2, lw=0.3)
end
for b in 3:3000
    plot!(X.(xs, b), Y.(xs, b); label="", c=2, lw=0.3)
end
plot!(X.(x,y), Y.(x, y); label="", c=1)
plot!(-X.(x,y), 2 .- Y.(x, y); label="", c=1, ls=:dot)
plot!(xlim=(-2.2, 2.2), ylim = (0, 2))
title!("y = √((x + 2)² + 2)")
```
