```julia
using Distributions
using StatsPlots
pyplot(fmt=:png, fontfamily="IPAPGothic",
    legendfontsize=12)
using KernelDensity
using Random
```

```julia
Random.seed!(464937311)
dist = Normal(170, 10)
n = 10
X = rand(dist, n)
X̄ = mean(X)
ikx = InterpKDE(kde(X))
plot(x -> 4pdf(dist, x), 120, 220; label="母集団", c=:red)
vline!([mean(dist)]; label="母平均", c=:red)
plot!(x -> pdf(ikx, x), 120, 220; label="標本", c=:blue)
vline!([X̄]; label="標本平均", c=:blue)
```
