提供された資料は、Datamethods Discussion Forumにおける、頻度論的な治療効果の統計結果の伝え方に関する議論の一部です。この議論では、従来の統計用語や解釈の問題点、より正確で分かりやすい表現方法が話し合われています。

Frankさんは議論の開始点（または初期の編集案）を提示し、P値は頻繁に誤用され、信頼区間の方がより多くの情報を含むため好ましいと述べています。また、「統計的有意性」や「非有意性」といった言葉を避け、「信頼区間」の誤解を防ぐために「データの値との相性が良い区間(compatibility interval)」という用語の使用を提案しています。

これに対し、SanderさんはFrankさんのいくつかの考え方に異論を唱えています。主な対立点は以下の通りです。

  1.  **P値の有用性について**:
  
    *   **Frankさんの考え（Sanderさんが異論を唱える点）**: P値は救いがたいものであり、利益よりも害をもたらす。したがって、P値はしばしば廃止されるべきである。特に大きなP値はほとんど情報を伝えない。
    
    *   **Sanderさんの異論**: P値自体が問題なのではなく、「有意性検定」として**誤用**されることが害の原因である。P値は信頼区間よりも賢明に廃止されるべきではない。P値は、仮説に対する**データの非互換性（不一致）の度合い**、つまり**相性の良さ**を連続的に測る尺度として捉えるべきである。P値は、その仮説に**反する情報量**（Shannonのサプライザル測度であるS値 log2(1/p)で測定）を測るものであり、p=0.4であれば1.3ビット、p=0.02であれば5.6ビットの情報量がある。P値は帰無仮説だけでなく、他の仮説についても計算されるべきである。単にP値の使用を放棄することは有用ではない。
<p></p>

  2.  **信頼区間（データの値との相性が良い区間）の評価について**:
  
    *   **Frankさんの考え**: 信頼区間はP値よりも多くの情報を含み、小さな効果の証拠を示すのに有用であるため、頻度論的な分析では好ましい。誤解を避けるために「データの値との相性が良い区間(compatibility interval)」という用語を使用することを提案している。
    
    *   **Sanderさんの異論**: 信頼区間は有意性検定よりも統計の乱用を防ぐ上で**優れていない**。信頼区間は同じ統計モデルに基づいた別表現に過ぎず、P値と同様に乱用されやすい。信頼区間は、誤解を招く0.05の二分法に依拠する欠点がある。特に「confidence」という言葉は**ひどい形容詞**であり、単にP値関数の要約に過ぎないものを「信頼」と表現することで**過信**を生む。信頼区間（データの値との相性が良い区間）は、P値 > 0.05となるすべてのパラメータ値の区間として、**単一のデータ分析に対しても解釈可能**である。これは、その区間内のパラメータ値に対して、データがlog2(1/0.05) = 4.3ビット未満の反証情報しか提供しないことを示す。この点において、SanderさんはFrankさんの「この単一のデータ分析に対して区間を解釈することはできず、長期的な特性としてのみ解釈できる」という記述が誤りであると述べている。
<p></p>

  3.  **統計的な問題の根本原因について**:
  
    *   **Frankさんの考え**: 統計的手法の**基礎に問題**があるため、多くのユーザーが誤用してしまう。もしある方法の基礎に根本的な問題があるために大多数がそれを誤用するなら、それは基礎が欠陥があるということである。
    
    *   **Sanderさんの異論**: 核となる問題は、統計手法自体ではなく、**誤った教え方**、**ユーザーの能力の限界**、そして分析や出版における**歪んだインセンティブ**といった**心理社会的な要因**によって引き起こされている。方法を置き換えるといった表面的な解決策は影響が非常に限定的である。
<p></p>

  4.  **頻度論とベイズ論の間の関係について**:
  
    *   **Frankさんの考え**: ベイズ論と頻度論の間のマッピング（等価性）は**特定の状況でのみ可能**であり、すべての状況で1対1の等価性があるわけではない。特にシーケンシャル検定、確率的なサンプルサイズN、複数のアウトカムといった状況では困難または不可能である。ベイズ法は全体として頻度論よりも**少ない仮定**をしている（特にサンプル空間に関する仮定がより考慮しやすいため）。
    
    *   **Sanderさんの異論**: そのような「特定の状況」は、Sanderさんが医学・健康分野で遭遇する**すべての問題**をカバーしている。すべての線形、ロジスティック、Cox回帰には少なくとも部分的なベイズの類似物があり、逆にすべてのベイズ分析はランダムパラメータ頻度モデルに直接マッピング可能である。このマッピングを無視したり忘れがちなのは、両極端な考え方をする人々（「extremists」）の心理社会的な理由であり、これが統計学を頻度論/ベイズ論という**分裂**に導いた。Frankさんの「ベイズは頻度論よりも少ない仮定をする」という主張は**全く間違っている**。ベイズ法は、**事前分布**という重要な仮定を追加し、サンプリングモデルの複雑さを隠蔽する。問題は統計的手法や数学的なものではなく、**提示の仕方**、つまり認知的なものである。
<p></p>

  5.  **事前分布の批判の必要性について**:
  
    *   **Frankさんの考え**: Sanderさんがいかなる事前分布の選択も疑問視する立場にいると感じており、これが根本的な意見の不一致である。Frankさんは、事前分布は「正しい」必要はなく、**合意される**必要があるだけだと信じている。事前分布は「表に出ている（plain site）」ものであり、批判可能である。
    
    *   **Sanderさんの異論**: すべての**関係者（stakeholder）**は、事前分布、サンプリングモデル（総モデル）、決定の根拠となる損失関数、そしてP値や事後確率などを生成するために使用された**あらゆる要素を疑問視する権利**がある。統計学者の仕事は、これらの要素を可能な限り正確、透明、明確にすることである。Sanderさんは、ベイズ法は事前分布を通じて分析を微妙に**操作する**大きな柔軟性をもたらす可能性があることを懸念している。ベイズ分析は、**文脈上のサンプリング物語の中で仮定をチェックしない**と危険である。

SanderさんとFrankさんは、「統計的有意性」や二分法的な解釈を避けるべきであるという点 や、誤りを公表することで改善を促すべきであるという点 など、多くの点で意見が一致しています。しかし、P値の価値、信頼区間の問題点、頻度論とベイズ論の間の関係性、そして統計的な問題の根本原因についての彼らの見解には明確な対立が見られます。特にSanderさんは、統計的な誤用の根本原因は方法論的なものではなく、言葉の選択や教育、出版システムなどの**人間的、社会的な側面**にあることを強調しています。
