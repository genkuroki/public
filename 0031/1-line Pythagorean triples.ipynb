```julia
f(n) = [(x, y, z) for x in 1:n for y in x:n for z in y:2n if x^2 + y^2 == z^2 && gcd(x, y) == 1]
@time a = f(1000)
@time a = f(1000)
@time a = f(1000)
print(a)
```

```julia
g(n) = [(x, y, isqrt(x^2+y^2)) for x in 1:n for y in x:n if (t = x^2 + y^2; t == isqrt(t)^2) && gcd(x, y) == 1]
@time b = g(1000)
@time b = g(1000)
@time b = g(1000)
print(b)
```
