```julia
using SpecialFunctions
mybinomial(n, k) = ((l, s) = logabsbinomial(n, k); s*exp(l))
```

```julia
?logabsbinomial
```

```julia
using Plots
default(fmt=:png, size=(400, 250))

n = 1000
k = 0:n
plot(k, mybinomial.(n, k); label="")
```

```julia
n = -3
k = 0:8
plot(k, mybinomial.(n, k); label="")
plot!(k, binomial.(n, k); label="", ls=:dash)
```
