```
MappedBy[x_] := x
MappedBy[x_, F___, G_] := MappedBy[x, F] // G

SetAttributes[TeXEq, HoldFirst]
TeXEq[x_] := TeXForm[HoldForm[x] == MappedBy[x, ReleaseHold, FullSimplify]]
TeXEq[x_, F__] := TeXForm[HoldForm[x] == MappedBy[x, ReleaseHold, F]]
TeXEqRaw[x_] := TeXForm[HoldForm[x] == MappedBy[x, ReleaseHold]]
```

```
Integrate[Log[Sin[x]], x] // TeXEq
```

$$
\operatorname{sinc}(x) =
\begin{cases}
\dfrac{\sin(x)}{x} & (x \ne 0) \\
1 & (x = 0) \\
\end{cases}
$$
