```julia
using Distributions

λ, α, τ = 3, 5, 5
@show cdf(Poisson(λ*τ), α-1)
@show ccdf(Gamma(α, 1/τ), λ);
```

```julia
using Distributions
using StatsPlots

α, τ = 5, 2
plot(λ -> cdf(Poisson(λ*τ), α-1), 0, 8;
    label="cdf(Poisson(λτ), α-1)")
plot!(λ -> ccdf(Gamma(α, 1/τ), λ), 0, 8; ls=:dash,
    label="ccdf(Gamma(α, 1/τ), λ)")
title!("α = $α,  τ = $τ")
```
