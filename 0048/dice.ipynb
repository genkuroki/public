```julia
using Distributions
using StatsPlots
default(fmt=:png)

dice = Categorical(6) # サイコロの確率分布
n = 10 # サイコロの個数
L = 10^5 # 繰り返しの回数
Y = zeros(L)
for i in 1:L
    Y[i] = sum(rand(dice, n)) # n個のサイコロの出目の和
end
histogram(Y; norm=true, alpha=0.3, bin=n-0.5:6n+0.5, label="data")
plot!(Normal(n*mean(dice), √n*std(dice)); label="normal approx.")
```
