```julia
using Distributions
using StatsPlots
default(fmt=:png)

dice = Categorical(6) # 
n = 10 # 
L = 10^5 # 
Y = zeros(L)
for i in 1:L
    Y[i] = sum(rand(dice, n)) # n
end
histogram(Y; norm=true, alpha=0.3, bin=n-0.5:6n+0.5, label="data")
plot!(Normal(n*mean(dice), n*std(dice)); label="normal approx.")
```
