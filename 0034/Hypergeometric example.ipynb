```julia
using Distributions
using StatsPlots
default(fmt=:png, size=(400, 250), titlefontsize=10)

k, N, n = 7, 10, 2
dist = Hypergeometric(k, N-k, n)
@show mean(dist), var(dist)
@show rationalize.((mean(dist), var(dist)))

xs = support(dist)
bar(xs, x -> pdf(dist, x); label="", title="$dist", xtick=xs)
```
