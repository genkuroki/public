```julia
using Distributions
using StatsPlots
default(fmt=:png, titlefontsize=12, legendfontsize=12)
```

```julia
n = 10^6
a, b,  = 6, 14, 3
X = rand(Gamma(a, ), n)
Y = rand(Gamma(b, ), n)
P = @. X/(X+Y)
U = @. X/Y;

stephist(P; norm=true, label="X/(X+Y)")
plot!(Beta(a, b); label="Beta(a,b)", ls=:dash)
title!("XGamma(a=$a, =$), YGamma(b=$b, =$)") |> display

stephist(U; norm=true, label="X/Y", xlim=(0, 2))
plot!(BetaPrime(a, b); label="BetaPrime(a,b)", ls=:dash)
title!("XGamma(a=$a, =$), YGamma(b=$b, =$)") |> display
```

```julia
n = 10^6
,  = 12, 28
X = rand(Chisq(), n)
Y = rand(Chisq(), n)
F = @. (X/)/(Y/)

stephist(F; norm=true, label="X/Y", xlim=(0, 2*/))
plot!(FDist(, ); label="FDist(, )", ls=:dash)
title!("XChisq(=$), YChisq(=$)")
```
